# 创建一个简单的线程

有两个线程在进程中执行，当主线程执行到`t=Thread()`时，创建了一个新的线程，执行到`s.start()`时，通知创建的新线程开始执行，此时当前进程中有两条线：主线程，子线程。

子线程执行work任务。
主线程继续向下执行。

由于输出控制台只有一个，两个线程都向这个控制台写内容，所以输出的显示才会很乱。

```python
from threading import Thread

def work(name):
    for i in range(100):
        print(f'{name} print {i}')
        
if __name__ == '__main__':
    t = Thread(target=work, args=('zhans',))
    t.start()

    for i in range(100):
        print('主线程', i)
```

# 线程池

只需要将任务给线程池，不需要操心线程执行结束一个任务，调度另一个任务。

```python
from concurrent.futures import ThreadPoolExecutor

def work(name):
    for i in range(100):
        print(f'{name} print {i}')
        
if __name__ == '__main__':
    # 1.先造线程池
    with ThreadPoolExecutor(16) as t:
        # 给线程池提交一个任务
        # t.submit(任务，执行任务时需要的参数，参数2。参数3)
        t.submit(work, '线程1')
        t.submit(work, '线程2')
        t.submit(work, '线程3')
```
