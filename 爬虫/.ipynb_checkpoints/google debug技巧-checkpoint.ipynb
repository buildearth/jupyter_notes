# 打断点之后查看js代码运行的结果

断点必须停在查看的位置

方式一：

在作用域中查看js代码中变量的值


方式二：

代码运行到想要看值的地方，即运行环境到当前作用域，在console中执行代码


方式三：
选中代码，鼠标停在上面查看值


# 查看函数声明的方式

注意：程序必须停在当前断点位置才能操作，右侧scope中显示的是当前断点是的变量值

方式一：鼠标悬停在函数上

方式二：打断点进入

点击进入，进入到函数中

方式三：在console中打印函数名，然后点击。程序必须停在当前行
