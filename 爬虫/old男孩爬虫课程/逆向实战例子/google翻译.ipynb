````python
import requests
from urllib.parse import urlencode
headers = {
    "user-agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/102.0.0.0 Safari/537.36",
    "referer": "https://translate.google.cn/",
}
session = requests.session()
resp = session.get('https://translate.google.cn/?sl=auto&tl=zh-CN&text=cc&op=translate')

import re
req_param_re = re.compile('"FdrFJe":"(?P<sid>.*?)".*"cfb2h":"(?P<lb>.*?)"')
# re.findall(req_param_re, d)
res = re.search(req_param_re, resp.text)
sid = res.group('sid')
lb = res.group('lb')
print(sid, lb)
print(session.cookies)

# 获取_reqid
'''
a = new Date,
pt = 3600 * a.getHours() + 60 * a.getMinutes() + a.getSeconds());
this.Wa = 1 + pt + 1E5 * this.wc;
'''
import datetime
from urllib.parse import urlencode
wc = 0
d = datetime.datetime.now()
pt = 3600 * d.hour + 60 * d.minute + d.second
start_reqid = int(1 + pt + 1E5 * wc)

# 发送第一个模拟的请求
url_data = {
    "rpcids": "MkEWBc",
    "source-path": "/",
    "f.sid": sid,
    "bl": lb,
    "hl": "zh-CN",
    "soc-app": "1",
    "soc-platform": "1",
    "soc-device": "1",
    "_reqid": int(start_reqid),
    "rt": "c",
}
send_url = 'https://translate.google.cn/_/TranslateWebserverUi/data/batchexecute?'
post_data = {
    'f.req': '[[["IjQUHc","[]",null,"generic"]]]'
}
send_url += urlencode(url_data)
resp = session.post(url=send_url, data=post_data)
print(resp.text)
print(session.cookies)
print(resp.url)
wc += 1
# _reqid = 

# 还需要再发一条模拟数据
url_data = {
    "f.sid": sid,
    "bl": lb,
    "hl": "zh-CN",
    "soc-app": "1",
    "soc-platform": "1",
    "soc-device": "1",
    "_reqid": start_reqid + wc * 100000,
    "rt": "j",
}
send_url = 'https://translate.google.cn/_/TranslateWebserverUi/browserinfo?'
post_data = {
    'f.req': '[9,1,1,[null,1080,1920],[null,937,615],[true,true,null,true],[false,0,2]]'
}
send_url += urlencode(url_data)
resp = session.post(url=send_url, data=post_data)
print(resp.text)
wc = wc+3

# 进行翻译
url_data = {
    "rpcids": "MkEWBc",
    "source-path": "/",
    "f.sid": sid,
    "bl": lb,
    "hl": "zh-CN",
    "soc-app": "1",
    "soc-platform": "1",
    "soc-device": "1",
    "_reqid": start_reqid + wc * 100000,
    "rt": "c",
}
send_url = 'https://translate.google.cn/_/TranslateWebserverUi/data/batchexecute?'
post_data = {
    'f.req': '[[["MkEWBc","[[\"funck\",\"auto\",\"zh-CN\",true],[null]]",null,"generic"]]]'
}
send_url += urlencode(url_data)
resp = session.post(url=send_url, data=post_data)
print(resp.text)
print(session.cookies)
print(resp.url)
````
