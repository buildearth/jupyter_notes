## python实现加密逻辑

````python
import requests
import base64
import json
from Crypto.Cipher import AES

def encrypt_one(data, key):
    aes = AES.new(key=key.encode('utf-8'), mode=AES.MODE_CBC, IV=b"0102030405060708")
    # data数据补齐
    
    mes = data.encode('utf-8')
    pad_len = 16 - len(mes) % 16
    mes += (pad_len * chr(pad_len)).encode("utf-8")
    mi = aes.encrypt(mes)
    return base64.b64encode(mi).decode()
    
    

def encrypt_data(data):
    # i，enc_sec_key这两个值要同时拷贝一个调试中的结果
    i = 'eC8zqKqbrqaqf49H'
    g = "0CoJUm6Qyw8W8jud"
    first = encrypt_one(data, g)
    enc_data = encrypt_one(first, i)
    enc_sec_key = '40dfe70c3fadcb5fe5f23b05ea55ad52a691cda7ce850c09eac73e8a475ac3743ebdf9532c347e1608f1a1772aeeeb671e372f3b8d1d9bdbfe8d0db10316e8301539fbde33e0b838753cb959122c65a4f685ede04dec4ad2b90b29e36effd9b3f23525ee6e86bafd63fda4c642de77cdb846d044a56802d84edd2c4e46c70dc3'
    
    return enc_data, enc_sec_key

url = 'https://music.163.com/weapi/song/enhance/player/url/v1?csrf_token='
headers = {
    "user-agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/102.0.5005.63 Safari/537.36",
}

# data = {"ids": '[65126]', "level": 'standard', "encodeType": 'aac', "csrf_token": ''}
params, encSecKey = encrypt_data('{"ids":"[65126]","level":"standard","encodeType":"aac","csrf_token":""}')

data = {
    "params": params,
    "encSecKey": encSecKey
}

resp = requests.post(url=url, headers=headers, data=data)
print(resp.text)
data = resp.json()['data']
print(data)

for music in data:
    url = music['url']
    id = music['id']
    # 保存音乐文件
    resp = requests.get(url=url)
    with open(f'{id}.m4a', mode='wb') as f:
        f.write(resp.content)
````
