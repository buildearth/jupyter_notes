### 打断点分析

r是在上面赋值的，打断点


再次点击翻译，进入到函数中


发现是MD5加密，不用看MD5的代码了，直接在python中使用MD5加密实现：
- t是MD5加密
- r 是时间戳
- i 是随机数
- sign 是MD5加密

navigator.appVersion的值：就是ua


## python模拟请求

````python
import requests
import time
import random
from hashlib import md5
headers = {
    'Referer': 'https://fanyi.youdao.com/',
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/102.0.5005.61 Safari/537.36',
}


def get_encrypt_data(word):
    navigator_appVersion = '5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/102.0.5005.61 Safari/537.36'
    bv_md5_obj = md5()
    bv_md5_obj.update(navigator_appVersion.encode('utf-8'))
    bv = bv_md5_obj.hexdigest()

    ts = str(int(time.time()*1000))  # python的时间戳转为js的实践戳
    # r + parseInt(10 * Math.random(), 10)  // 将随机小数乘以10，转为10进制
    salt = ts + str(random.randint(0, 10))
    sign_md5 = md5()
    # sign: n.md5("fanyideskweb" + e + i + "Ygy_4c=r#e#4EX^NUGUc5")
    sign_str = "fanyideskweb" + word + str(salt) + "Ygy_4c=r#e#4EX^NUGUc5"

    # sign_str = "fanyideskweb牛啊\n16539245231903Ygy_4c=r#e#4EX^NUGUc5"
    sign_md5.update(sign_str.encode('utf-8'))
    sign = sign_md5.hexdigest()

    return {"bv": bv, "ts": ts, "salt": salt, "sign": sign}

word = "牛啊"
encrypt_data = get_encrypt_data(word)

data = {
    "i": word,
    "from": "AUTO",
    "to": "AUTO",
    "smartresult": "dict",
    "client": "fanyideskweb",
    "salt": encrypt_data["salt"],
    "sign": encrypt_data["sign"],
    "lts": encrypt_data["ts"],
    "bv": encrypt_data["bv"],
    "doctype": "json",
    "version": "2.1",
    "keyfrom": "fanyi.web",
    "action": "FY_BY_REALTlME",
}

url = "https://fanyi.youdao.com/translate_o?smartresult=dict&smartresult=rule"
cookie = {
    "OUTFOX_SEARCH_USER_ID": '1689272698@10.108.160.105',
    "OUTFOX_SEARCH_USER_ID_NCOO": '144072625.36730197',
    "fanyi-ad-id": '305838',
    "fanyi-ad-closed": '1',
    "___rl__test__cookies": str(int(time.time()*1000)),
}
resp = requests.post(
    url=url,
    data=data,
    headers=headers,
    cookies=cookie,
)

print(resp.json())
````
