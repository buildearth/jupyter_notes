substr传递一个参数，是从参数位置取到末尾


````python
# 函数_0x9843d3
def fn(a, b, c):
    if 0 == b:
        return a[c:]
    res = a[:b]  # a.substr(0, b)  0, 0+b
    res += a[b+c:]
    return res 

fn('abcdefghigklmn', 2, 4)
````
