# mangodb


是一个非关系型数据库，非常适合超大数据集的存储。

查询的效率会很高。

## 安装

下载地址：https://www.mongodb.com/try/download/community

将mongodb目录下的bin文件夹添加到环境变量

## 使用

创建数据库->集合collection(放一个一个的数据)->文档document(一个文档就是一个数据)

不用管数据的类型

命令
```
show dbs  显示所有数据库
db 查看当前使用的数据库

use python  创建并使用数据库，只是内存层面的数据库（如果存数据了，才会在硬盘上创建数据库）

db.createColloction("person")  创建collection,db是当前数据库。

db 是一个代称，永远是当前数据库

db.dropDatabase()  # 删除当前数据库

show collections 查看当前数据库中所有集合

db.stu.insert({"name": "xxx"})  创建集合并插入一个数据(集合再直接插入，集合不在新建集合再插入)

db.集合名.drop()  删除集合


```

## 增删改查

```
插入多条
    db.stu.insert({name:"xxx"})
    db.stu.insert([{name:'xc'},{name:'yy'}])
    pymongo新版本中改版了：
        insertOne,insertMany两个
        

```

查询
db.stu.find()  所有数据

db.stu.find({age:33})  {}中放的是条件, age为33的数据

db.stu.find({age:{$gt:33}})  年龄大于33

db.stu.find($and:[{age:38}, {name:'xx'}])  多条件and

一般不会使用复杂查询，多用于简单查询，数据全整出来的情况。

db.stu.findOne()
