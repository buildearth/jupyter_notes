当使用selenium工具时，浏览器很清楚的知道是通过selenium来启动的，会在浏览器中设定一些参数来标识当前浏览器是通过selenium.

使用selenium自动化网页时，有一定的概率会被目标网站识别，一旦被检测到，目标网站会拦截该客户端做出的网页操作。

每一个浏览器访问网站时，都会带上特定的指纹特征，网站会解析这些特征，从而判断这次访问是不是自动化程序

一个最广为人知的特征是window.navigator.webdriver，该特征直接标明此浏览器是webdriver程序。当一个浏览器通过selenium启动后，在开发者工具中输入这个属性，会发现被标为 true， 而手工打开的浏览器是 false。

如：12306的登录验证

如果是selenium打开并且想去通过事件链滑动，永远是不会成功的，因为12306服务器检测出浏览器是通过selenium打开的。

# 第一种跳过selenium检测的方式

在给selenium设置一个参数

```python
option = Options()
# 可以去掉显示的那个"自动化工具xxxx"
option.add_experimental_option('excludeSwitches', ['enable-automation'])
# 它可以取消掉webdriver，可以让服务器无法识别到你是通过selenium打开的参数之一
option.add_argument('--disable-blink-features=AutomationControlled')

web = Chrome(options=option)
```
