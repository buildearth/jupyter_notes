# 网站搜索的问题

问题描述：

在网页上看搜索框搜索请求数据时，会将输入的内容编码然后在发送请求


在看网页代码时，看到拼接url的代码如下：用escape对`摩托车`三个字进行了编码


最开始的解决方案是：
- 在后端用python模拟excape对字符进行编码
- 使用request.get发送请求

页面请求不到搜索的内容，其原因是python模拟好请求后，url中有`%`这个特殊字符，requests在请求时会将这个特殊字符转为`%25`，传递到网页的服务器时，解码时，解码错误，找不到对应得数据。

最终解决方案：
- 将url中得keyword的带%的数据替换成中文
- 发送requests请求，可以拿到正确的数据，因为requests会自动处理编码

两个url对比：都可以拿到数据

在后续request请求时，这个参数直接写要搜索的内容即可，requests会自动处理。
