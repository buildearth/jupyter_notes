## 解析科目打开的页面-拿到所有的考试文件名字和其所在文件目录

在科目页面中，真正的考试隐藏在最内层，外层有可能是目录，也有可能是考试题。由外到最层的考试题的层级深度是不同的，有的是两级、有的是三级、有的当前就是。


所有由外向里找，不知道有多少层，用递归来找到最后一层，不建议使用递归，会导致不可预估的问题，当你的递归写的不正确时。


仔细观察页面的结构：

未展开层级


展开第一个层级到第一个最内层


由外向内寻找最内层不知道层级深度，不好找，所以反向思路，由内向外推，先找到最内层的(就是考试题)标签，再由内向外推，找到外层的每一层，这样层级结构就出来了。

xpath的新知识点：从当前节点向外层查找 `section_point_item.xpath("./ancestor::ul[@class='section-item' or @class='chapter-item']")  `

拿到考试文件的目录结构，方便存储。

## 解析json数据，并进行保存

解析出json中的题目、选项、答案、解析。

这个网站上的上的题型有：
- 单选题
- 多选题
- 判断题

- 材料题
    - 材料题下，有单选、多选、判断、简述这几类。
- 简述题

材料题和其他的题型在json中的位置不一样；
- 材料题，题在materials中

- 其他题，题在questions中


解析代码
```python
    def parse_question(self, resp, **kwargs):
        # print(resp.json())
        dir_path = resp.meta["dir_path"].strip().replace("\t", "").replace("\n", "")
        file_name = resp.meta["file_name"].strip().replace("\t", "").replace("\n", "")
        datas = resp.json()["Data"]
        for data in datas:
            questions = data.get("questions", None)
            if questions:
                for question in questions:
                    question_info = self.process_question(question)
                    # print(question_info)
                    yield {
                        "dir_path": dir_path,
                        "file_name": file_name,
                        "question_info": question_info
                    }
            else:
                # 材料题
                materials = data.get("materials")
                for material in materials:
                    material_content = material["material"]["content"]
                    q_list = []
                    # 材料题下的所有题：有单选、多选、判断、简述
                    for q in material["questions"]:
                        q_info = self.process_question(q)
                        q_list.append(q_info)

                    qs = material_content + '\n\n' + "\n".join(q_list)
                    yield {
                        "dir_path": dir_path,
                        "file_name": file_name,
                        "question_info": qs
                    }


    def process_question(self, question):
        content = question["content"]
        options = question["options"]
        analyse = question["textAnalysis"]
        opt_list = []
        res_list = []
        for opt in options:
            opt_content = opt["content"]
            opt_name = opt["name"]
            is_right = opt["isRight"]
            opt_list.append(f'{opt_name}.{opt_content}')
            if is_right == 1:
                # 该选项是正确答案, 可能是多选，单选，判断题， 选择题和判断题的name值不一样，可以进行区分
                # - 单选多选，正确答案是存储的： ABCDEFG...等一些字母的选项
                # - 判断题，正确答案存储的是： 对 错
                if opt_name in "ABCDEFGH":
                    res_list.append(opt_name)
                else:
                    res_list.append(opt_content)
        # 给管道保存数据
        question_info = "{content}\n{opts}\n正确答案：{res}\n解析:{analyse}".format(
            content=content, opts="\n".join(opt_list), res=','.join(res_list), analyse=analyse)
        return question_info
```

在pipline中存储题
```python
import os

class ZhogdawxPipeline:
    def process_item(self, item, spider):
        dir_path = item['dir_path']
        file_name = item['file_name']
        question_info = item['question_info']
        if not os.path.exists(dir_path):
            os.makedirs(dir_path)
        # 拼接存储路径
        file_path = os.path.join(dir_path,f"{file_name}.md")
        with open(file_path, mode='a', encoding='utf-8') as f:
            f.write(question_info)
        return item
```
