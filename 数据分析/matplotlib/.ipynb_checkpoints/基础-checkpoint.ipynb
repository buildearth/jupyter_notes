是一个强大的Python绘图和数据可视化的工具包

安装： pip install matplotlib

引用： import matplotlib.pyplot as plt

绘图函数（折线图）： plt.plot()

显示图像：plt.show()

````python
import matplotlib.pyplot as plt
plt.plot()  # 将给定的点连起来，第一个参数是x第二个参数是y,第三个参数是个字符串，指定画图的类型
plt.show()
````

````python
plt.plot([1,2,3,4], [2,3,5,5])
plt.show()
````

````python
plt.plot([1,2,3,4], [2,3,5,5],"yo-")
plt.show()
````

# 函数周边

在一个图上绘制多条线，写多个plot即可。

```python
# 设置图像标题
plt.title("标题")

# 设置x轴名称
plt.xlabel()

# 设置y轴名称
plt.ylabel()

# 设置x轴范围
plt.xlim(0,5)

# y轴范围
plt.llim()

# x轴刻度
plt.xticks([0,2,4])

# y轴刻度
plt.yticks()

# 设置曲线图例,描述线是哪条
plt.lehend()



```

````python
plt.plot([1,2,3,4], [2,3,5,5],"yo-", label="line1")
plt.plot([2,2,3,5], [4,6,7,8],"--", label="line2")
plt.legend()  # 在图中显示label的信息，也可以传递plt.plot的变量来显示
plt.show()
````

````python
plt.xticks([0,2,4])
plt.show()
````

pandas 和 matplotlib结合

````python
import pandas as pd
import numpy as np
````

````python
df = pd.DataFrame({'one':[1,2,3], 'two':[4,5,6]}, index=['a','b','c'])
print(df)
df.plot()
plt.show()
````

````python
x = np.linspace(0,10, 11)
plt.plot(x, label="y=x")
plt.plot(x,x**2, label="y=y^2")
plt.plot(x,(3*x**3), label="y=3*x^3")
plt.ylim(0,500)
plt.legend()
plt.show()
````

````python
fig = plt.figure()
ax1 = fig.add_subplot(2,2,1)
ax1.plot([1,2,3],[5,6,7])
ax2 = fig.add_subplot(2,2,2)
ax2.plot([1,2,3],[11,22,33])
plt.show()
````
